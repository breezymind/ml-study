```python
from konlpy.utils import pprint
from gensim.models import word2vec
```

```python
model = word2vec.Word2Vec.load('./model/daily_news_test')
```

```python
pprint(model.index2word[:20])
```

```python
vec = model.syn0
vocab = model.vocab
```

```python
import numpy as np
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, random_state=0)
np.set_printoptions(suppress=True)
two_d_embeddings = tsne.fit_transform(vec)
```

```python
import matplotlib.pyplot as plt
import pylab as Plot

Plot.figure(figsize=(30,30), dpi=20)
```

```python
plt.scatter(two_d_embeddings[:,0], two_d_embeddings[:,1], 10)
```

```python
tmp = {}

idx = 0
for label in vocab.keys():
    tmp[label] = two_d_embeddings[idx]
    idx+=1
```

```python
for row in tmp:
    Plot.annotate(row, (tmp[row][0],tmp[row][1]))
```

```python
plt.show()
```
